### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
# Votre code ici
setosa_petals = features[:50, 2]
versicolor_petals = features[50:100, 2]
virginica_petals = features[100:, 2]

means = [np.mean(setosa_petals), np.mean(versicolor_petals), np.mean(virginica_petals)]

ecart_types = [np.std(setosa_petals), np.std(versicolor_petals), np.std(virginica_petals)] 

plt.figsize=(10, 6)
species = ['setosa', 'versicolor', 'virginica']

plt.bar(species, means, color=['red', 'blue', 'green'])
plt.errorbar(species, means, yerr=ecart_types, fmt='o', color='black')
plt.xlabel('espece')
plt.ylabel('petal_length')
plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
# Votre code ici
plt.figsize=(10, 6)

cumsum_setosa = np.cumsum(features[:50, 3])
cumsum_versicolor = np.cumsum(features[50:100, 3])
cumsum_virginica = np.cumsum(features[100:, 3])

plt.plot(cumsum_setosa, color='red', label='setosa')
plt.plot(cumsum_versicolor, color='blue', label='versicolor')
plt.plot(cumsum_virginica, color='green', label='virginica')

plt.title('cumulative sum of petal_width')
plt.xlabel('nombre de fleurs')
plt.ylabel('cumulative somme')
plt.legend()
plt.grid()
plt.show()
```

# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
# Votre code ici
correlations = np.corrcoef(features, rowvar=False) 
print(correlations)
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
# Votre code ici
plt.figsize=(10, 6)


plt.imshow(correlations, cmap='hot', interpolation='nearest', vmin=0, vmax=1)
noms = ['petal_length', 'petal_width', 'sepal_length', 'sepal_width']
plt.title('matrice de correlation')
plt.xticks(np.arange(4), noms)
plt.yticks(np.arange(4), noms)
plt.colorbar(label = 'correlation')
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici
var_setosa = np.sum(np.var(features[:50, 3]))
var_versicolor = np.sum(np.var(features[50:100, 3]))
var_virginica = np.sum(np.var(features[100:, 3]))

print('variance setosa: ', var_setosa)
print('variance versicolor: ', var_versicolor)
print('variance virginica: ', var_virginica)
print('Espece avec grande variance est : virginica')

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('histogramme setosa')
plt.hist(features[:50, 0], alpha=0.5, color='red', label='sepal_length setosa')
plt.hist(features[:50, 1], alpha=0.5, color='blue', label='sepal_width setosa')
plt.hist(features[:50, 2], alpha=0.5, color='green', label='petal_length setosa')
plt.hist(features[:50, 3], alpha=0.5, color='yellow', label='petal_width setosa')
plt.xlabel('caracteristiques')
plt.ylabel('frequence')
plt.legend()

plt.subplot(1, 3, 2)
plt.title('histogramme versicolor')
plt.hist(features[50:100, 0], alpha=0.5, color='red', label='sepal_length versicolor')
plt.hist(features[50:100, 1], alpha=0.5, color='blue', label='sepal_width versicolor')
plt.hist(features[50:100, 2], alpha=0.5, color='green', label='petal_length versicolor')
plt.hist(features[50:100, 3], alpha=0.5, color='yellow', label='petal_width versicolor')
plt.xlabel('caracteristiques')
plt.ylabel('frequence')
plt.legend()

plt.subplot(1, 3, 3)
plt.title('histogramme virginica')
plt.hist(features[100:, 0], alpha=0.5, color='red', label='sepal_length virginica')
plt.hist(features[100:, 1], alpha=0.5, color='blue', label='sepal_width virginica')
plt.hist(features[100:, 2], alpha=0.5, color='green', label='petal_length virginica')
plt.hist(features[100:, 3], alpha=0.5, color='yellow', label='petal_width virginica')
plt.xlabel('caracteristiques')
plt.ylabel('frequence')
plt.legend()
```
